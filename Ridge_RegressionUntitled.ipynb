```python
import numpy as np
import pandas as pd
```

### Loading the data

```python
df = pd.read_csv('Advertising.csv', index_col = 0)
df.head()
```

### Creating the features(X) and labels(y)

```python
X = df.drop(columns = 'Sales')
```

```python
y = df['Sales']
```

### Creating polynomial features

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_conveter = PolynomialFeatures(degree = 3, include_bias = False)
```

```python
poly_features = poly_conveter.fit_transform(X)
poly_features.shape
```

### Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

### Scaling the data (STANDARDIZATION)


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train) # Use X_train
```

```python
X_train = scaler.transform(X_train)
X_train
```

```python
X_test = scaler.transform(X_test)
```

### RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=10) # run basically like a regression model
```

```python
ridge_model.fit(X_train, y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
test_predictions
```

TESTING PERFORMANCE

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MAE 
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
RMSE
```

### Now to check if alpha or lambda of 10 is good or not we need to check other alpha numbers too, for this use RidgeCV

```python
from sklearn.linear_model import RidgeCV
```

```python
ridge_cv_model = RidgeCV(alphas = (0.1, 1, 10), scoring = 'neg_mean_absolute_error')
```

```python
ridge_cv_model.fit(X_train, y_train)
```

```python
ridge_cv_model.alpha_ # Returns which alpha performed the best
```

```python
test_predictions = ridge_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
RMSE
```
