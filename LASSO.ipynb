```python
import numpy as np
import pandas as pd
```

### Loading the data and creating features(X) and labels (y)

```python
df = pd.read_csv('Advertising.csv', index_col=0)
```

```python
X = df.drop(columns = 'Sales')
X.head()
```

```python
y = df['Sales']
```

### Creating polynomial features

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_converter = PolynomialFeatures(degree = 3, include_bias=False)
```

```python
poly_features = poly_converter.fit_transform(X)
```

### Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

### Standarization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

## LASSO

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv_model = LassoCV(eps = 0.1, n_alphas=100)
```

```python
lasso_cv_model.fit(X_train, y_train)
```

```python
lasso_cv_model.alpha_
```

```python
test_preditcions = lasso_cv_model.predict(X_test)
```

### TESTING

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, test_preditcions)
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_preditcions))
RMSE
```
