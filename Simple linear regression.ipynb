```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the data set

```python
df = pd.read_csv('Advertising.csv', index_col=0)
df['Total_spend'] = df['TV'] + df['Radio'] + df['Newspaper'] 
df.head()
```

```python
plt.figure(figsize = (5, 3), dpi = 200)
sns.regplot(data = df, x = 'Total_spend', y = 'Sales')
```

### Manually getting the regression line

```python
X = df['Total_spend']
y = df['Sales']
```

```python
# y = mx + b
# y = B1x + B0
# help(np.polyfit)
```

```python
np.polyfit(X, y, deg = 1)
#  firt shows the B1 then B0
```

```python
potential_spend = np.linspace(0, 500, 100)
```

```python
predicted_sales = 0.04868788 * potential_spend + 4.24302822
```

```python
sns.scatterplot(data = df, x = 'Total_spend', y = 'Sales')
plt.plot(potential_spend, predicted_sales, color = 'red')
```

```python
spend = 200
sales = 0.04868788 * spend + 4.24302822
sales
```

#### now lets do a 3 degree polyfit


```python
np.polyfit(X, y, deg = 3)
```

```python
#  y = B3 * x**3 + B2 * x**2 + B1 * x + B0
```

```python
spend = np.linspace(0, 500, 100)
pred_sales =  3.07615033e-07 * spend**3 +  -1.89392449e-04 * spend**2 +  8.20886302e-02 * spend +  2.70495053e+00
```

```python
plt.plot(spend, pred_sales, color = 'red')
sns.scatterplot(data = df, x = 'Total_spend', y = 'Sales')
#  Now we have two models one with a straight line and this one:
```
